INTRODUCTION

In the era of rapid urbanization, cities have become complex "living organisms" that consume vast amounts of resources and exhale significant waste. This project explores the Urban Metabolism of New York City by conducting a comprehensive audit of its built environment. Using a high-dimensional dataset sourced from the U.S. Government (Data.gov), we analyze the energy and water "nutrients" that flow into NYC's thousands of buildings and the resulting Greenhouse Gas (GHG) emissions that contribute to its carbon footprint.
​Leveraging the power of Python (Pandas, NumPy, Matplotlib,Plotly and Seaborn), this study aims to transform raw benchmarking data into a strategic resource audit. By examining 257 distinct attributes—ranging from building age and occupancy to fuel oil usage and GHG intensity—this project seeks to quantify the metabolic efficiency of New York City and identify paths toward a more sustainable, low-carbon future.

PROBLEM STATEMENT

New York City's goal to achieve carbon neutrality by 2050 is challenged by an aging and diverse building stock that accounts for nearly 70% of the city’s total emissions. While the city collects extensive data via Local Law 84, the raw data is often "noisy" and fragmented, making it difficult for policymakers and engineers to see the true picture of resource waste.

The primary problems addressed in this data-driven audit are:

* *Data Sparsity and Feature Overload:* The presence of 265 initial columns, many filled with "Not Available" strings, obscures critical KPIs. We must perform rigorous Feature Selection to focus on the 257 columns that provide the highest informational value.
* *Incomplete Metabolic Profiles:* Gaps in water and energy reporting create "blind spots" in the city's audit. This project utilizes Statistical Imputation (Median and Mode) to create a complete dataset for accurate modeling.
* *Carbon Intensity Hotspots:* Without advanced Multivariate Analysis, the specific building types or geographical boroughs that act as "carbon hotspots" remain hidden.

<h2>2.5 Gini Coefficient Analysis</h2>

```python
def gini(x):
    # Ensure input is a numpy array and remove any potential negatives
    x = np.array(x)
    if np.any(x < 0):
        x -= np.min(x)
    # The Gini calculation logic
    x = np.sort(x)
    n = len(x)
    index = np.arange(1, n + 1)
    return ((np.sum((2 * index - n - 1) * x)) / (n * np.sum(x)))
```

```python
# Convert to numeric and drop NaNs to ensure the function gets clean numbers
site_eui_data = pd.to_numeric(df_feature_selected['Site EUI (kBtu/ft²)'], errors='coerce').dropna()

# Now run the function
site_eui_gini = gini(site_eui_data)
print(f"Gini Coefficient for Site EUI: {site_eui_gini:.4f}")
```

<h2>Gini Coefficient Insights: Site EUI (kBtu/ft²)</h2>

<p>The <strong>Gini coefficient</strong> measures inequality or concentration for a single variable. Here, it is applied to <strong>Site EUI</strong>, which represents energy use per square foot of a building.</p>

<p><strong>Value Range:</strong></p>
<ul>
    <li><strong>0</strong> – perfect equality (all buildings use energy at the same intensity.</li>
    <li><strong>1</strong> – maximum inequality (one building consumes all energy relative to its size, others almost none.</li>
</ul>

<p><strong>Result for NYC Dataset:</strong></p>
<ul>
    <li>Gini Coefficient for <strong>Site EUI</strong>: <strong>0.9881</strong></li>
</ul>

<p><strong>Interpretation:</strong></p>
<ul>
    <li>This extremely high Gini coefficient indicates a <strong>very uneven distribution</strong> of energy intensity across NYC buildings.</li>
    <li>Most buildings have relatively low energy intensity, but a few buildings (likely large commercial complexes or specialized facilities like power stations) have extremely high energy intensity, skewing the distribution.</li>
    <li>This highlights the need for targeted energy efficiency measures for the high-intensity buildings.</li>
</ul>

TITLE: ANALYSIS OF PROPERTY SIZE INEQUALITY

```python
# 10. Lorenz Curve: Property GFA
# -------------------------------
def lorenz_curve(values):
    sorted_vals = np.sort(values.dropna())
    cumvals = np.cumsum(sorted_vals)
    lorenz = np.insert(cumvals / cumvals[-1], 0, 0)
    return lorenz

gfa_vals = filter_outliers(df['Property GFA - Self-Reported (ft²)'])
lorenz = lorenz_curve(gfa_vals)
n = len(gfa_vals)
plt.figure(figsize=(7,7))
plt.plot(np.linspace(0,1,n+1), lorenz, color='blue', label='Lorenz Curve')
plt.plot([0,1],[0,1], color='red', linestyle='--', label='Equality Line')
plt.title('Lorenz Curve - Property GFA', fontsize=16)
plt.xlabel('Cumulative Share of Buildings')
plt.ylabel('Cumulative Share of GFA')
plt.legend()
plt.tight_layout()
plt.show()
```

​NAME: LORENZ CURVE (PROPERTY GFA)

TITLE: MULTI-DIMENSIONAL PERFORMANCE COMPARISON BY BOROUGH

```python
# Metrics per borough
metrics = ['Total (Location-Based) GHG Emissions (Metric Tons CO2e)',
           'Site EUI (kBtu/ft²)',
           'Property GFA - Self-Reported (ft²)',
           'ENERGY STAR Score',
           'Water Use (All Water Sources) (kgal)']

# Group by borough and compute average
df_borough = df.groupby('Borough')[metrics].mean().dropna()

# Normalize 0-1 per metric
df_norm = (df_borough - df_borough.min()) / (df_borough.max() - df_borough.min())

# Radar setup
categories = list(df_norm.columns)
N = len(categories)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # close the circle

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

# Plot each borough
for idx, row in df_norm.iterrows():
    values = row.tolist()
    values += values[:1]  # close circle
    ax.plot(angles, values, linewidth=2, label=idx)
    ax.fill(angles, values, alpha=0.25)

# Add labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.set_yticks([0.25, 0.5, 0.75, 1.0])
ax.set_yticklabels(['0.25','0.5','0.75','1.0'])
plt.title('Normalized Radar Chart by Borough', fontsize=16)
plt.legend(loc='upper right', bbox_to_anchor=(1.2,1.1))
plt.show()
```
